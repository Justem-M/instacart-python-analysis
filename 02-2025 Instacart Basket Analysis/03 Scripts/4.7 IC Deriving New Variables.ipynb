```python
# Creating subst
df = df_ords_prods_merge[:1000000]
```

```python
# Checking subset
df.shape
```

```python
# If-statement with user-defined functions - price_label

def price_label(row):
    
    if row ['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15: 
        return 'High range'
    else: return 'Not enough data'
```

```python
# Use function on df
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# checking applied function
df.head()
```

```python
# Checking values in new column
df['price_range'].value_counts()
```

```python
# Checking max price
df['prices'].max()
```

## 6. Task

### 6.1 Busiest Days

```python
# Frequency orders_day_of_week - identifying busiest days
df_ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

0 and 1 are the two busiest days
3 and 4 are the two slowest days
6, 2 and 5 are regularly busy

```python
# Creating column with values for "busiest days"

result2 = []

for value in df_ords_prods_merge["orders_day_of_week"]:
    if value == 0 or value == 1:
        result2.append("Busiest days")
    elif value == 4 or value == 3:
        result2.append("Least busy days")
    else:
        result2.append("Regularly busy days")
```

```python
result2
```

```python
# Combine with dataframe
df_ords_prods_merge['busiest_days'] = result2
```

```python
df_ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

```python
df_ords_prods_merge.head()
```

All categories present. Seems to be much fewer least busy days. Regularly busy days have the most entries, which makes sense. Distribution makes sense. No missing or incorrect values in this column.

### 6.2 Busiest Hours

```python
# Frequency orders_hoour_of_day - identifying busiest hours
df_ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

Busiest hours - 10 am till 4 pm
Least busy hours - 0 to 6 am
Regular hours - 7 am to 9 am and 5 pm to 11 pm

```python
# Creating column with values for "busiest hours"

result3 = []

for value in df_ords_prods_merge["order_hour_of_day"]:
    if 10 <= value <= 16:
        result3.append("Busiest hours")
    elif 0 <= value <= 6:
        result3.append("Least busy hours")
    else:
        result3.append("Regularly busy hours")
```

```python
result3
```

```python
# Combine with dataframe
df_ords_prods_merge['busiest_hours'] = result3
```
