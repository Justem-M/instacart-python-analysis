## 3. Charts - Plots

### 3.1 Bar Charts

```python
# Create bar chart
df['orders_day_of_week'].value_counts().plot.bar()
```

```python
# Create bar chart with original order and making it an object
bar = df['orders_day_of_week'].value_counts().sort_index().plot.bar()
```

```python
# Create bar chart with original order and making it an object and changing the colours
bar_c = df['orders_day_of_week'].value_counts().sort_index().plot.bar(color =['purple', 'blue', 'green', "yellow"])
```

```python
# Some extra - using seaborn's colour palettes
# Picking a palette
palette = sns.color_palette('colorblind', as_cmap=False)
```

```python
# Now plotting with that palette
bar_c = df['orders_day_of_week'].value_counts().sort_index().plot.bar(color=palette)
```

```python
# Adding labels and title
plt.figure(figsize=(10,6))
bar_c = df['orders_day_of_week'].value_counts().sort_index().plot.bar(color=palette)
plt.title('Orders by Day of Week', fontsize=16)
plt.xlabel('Day of the Week', fontsize=12)
plt.ylabel('Number of Orders', fontsize=12)
plt.show()
```

```python
# Save bar to PC
bar_c.figure.savefig(os.path.join(path, '04 Analysis', 'Visualisations', 'bar_orders_dow.png'))
```

### 3.3 Line Charts and Sampling

```python
# Creating seed for sample
np.random.seed(4)
dev = np.random.rand(len(df)) <= 0.7
```

```python
# Check list
dev
```

```python
# Splitting dev in two groups
big = df[dev]
small = df[~dev]
```

```python
# Checking the size of the dfs
len(df)
```

```python
len(big) + len(small)
```

```python
# Reducing to necessary columns only
df_2 = small[['orders_day_of_week', 'prices']]
```

```python
# Creating line plot
line = sns.lineplot(data = df_2, x = 'orders_day_of_week', y = 'prices')
```

```python
# Saving line chart to PC
line.figure.savefig(os.path.join(path, '04 Analysis', 'Visualisations', 'line_prices_dow.png'))
```

### 4.4 Customer Demographics - Connection between Age and Family Situation

```python
# Reducing to necessary columns only
df_4 = small[['age', 'num_dependants']]
```

```python
# Creating line plot
line_3 = sns.lineplot(data = df_4, x = 'age', y = 'num_dependants')
```

The values fluctuate slightly, but there is no clear trend, indicating a consistent relationship between age and number of dependents. Most age groups appear to have an average of around 1.5 dependents, with only minor variations. 
The line chart shows a high variability from one age to the next, probably because of the sample size. If needed, one could aggregate ages into age groups to clarify patterns. 

```python
# Saving line chart to PC
line_3.figure.savefig(os.path.join(path, '04 Analysis', 'Visualisations', 'line_age-num_dependants.png'))
```
